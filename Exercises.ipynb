```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Titanic = sns.load_dataset('titanic')
```

```python
Titanic.head()
```

```python
Titanic = Titanic.dropna(subset = ['fare','age'])
```

```python
sns.jointplot(x='fare',y='age',data = Titanic)
plt.show()
```

```python
Titanic.head()
```

```python
sns.histplot(x = 'fare', data = Titanic, color = 'red', bins = 18)
plt.xlim(0,Titanic['fare'].max())
plt.show()
```

```python
Titanic['fare'].max()
```

```python
sns.boxplot(x = 'class', y = 'age', data = Titanic, palette = ['deepskyblue','lightgreen','orange'])
```

```python
sns.swarmplot(x='class',y = 'age', data = Titanic, palette = ['lightgreen','orange','skyblue'])
```

```python
sns.countplot(x = 'sex', data=Titanic, palette = ['skyblue','seagreen'])
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(Titanic.corr(numeric_only = True), cmap = 'coolwarm', center = 0)
plt.show()
```
